```python
import pandas as pd
import matplotlib.pyplot as plt

data_path = '../data/processed/processed.csv'
df = pd.read_csv(data_path, parse_dates=['date'])

print(df.head())
print(df.describe())

plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['Gulf'], label='Gulf Freight Price')
plt.title('Gulf Freight Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()


```
